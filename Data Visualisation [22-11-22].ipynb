```python
## Outlier Detection
### 1: When the data is Normally Distribution or Close to
###    Normally Distribution
### 2: Box PLot When Data Is Skewed
### 3: 
```

```python
tips.head()
```

```python
plt.boxplot(tips["tip"]);
```

```python
sns.distplot(tips["tip"]);
```

```python
tips["tip"].skew()
```

```python
p25=tips["tip"].quantile(0.25)
p25
```

```python
p75=tips["tip"].quantile(0.75)
p75
```

```python
iqr=p25-p75
iqr
```

```python
upper_limit=p75+(1.5*iqr)
upper_limit
```

```python
lower_limit=p25-(1.5*iqr)
lower_limit
```

```python
len(tips[tips["tip"]< upper_limit])
```

## How To Handle Outliers :-
    1- Trimming : not recommended as info is lost
    2- Capping  : Data Balance with Upper limit or Lower limit 

```python
dff=tips[(tips["tip"]<lower_limit)&(tips["tip"]>upper_limit)]
```

```python
dff["tip"].skew()
```

```python
new_df_cap = tips.copy()  ## create a copy for handling outlier in tips
new_df_cap


new_df_cap['tip'] = np.where(new_df_cap['tip'] > upper_limit, upper_limit, 
                                              np.where(new_df_cap['tip'] < lower_limit, lower_limit,
                                                       new_df_cap['tip'])) # for middle part of box which can be replaced with any one
```
